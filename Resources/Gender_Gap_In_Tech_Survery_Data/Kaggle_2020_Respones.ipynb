```python
import pandas as pd

df = pd.read_csv("kaggle_survey_2020_responses.csv")
df.head()
```

```python
# Way too many columns
df = df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6','Q24', 'Q37_Part_1', 'Q37_Part_2', 'Q37_Part_3', 'Q37_Part_4', 'Q37_Part_5', 'Q37_Part_6', 'Q37_Part_7', 'Q37_Part_8', 'Q37_Part_9', 'Q37_Part_10', 'Q37_Part_11', 'Q7_Part_1', 'Q7_Part_2', 'Q7_Part_3', 'Q7_Part_4', 'Q7_Part_5', 'Q7_Part_6', 'Q7_Part_7', 'Q7_Part_8', 'Q7_Part_9', 'Q7_Part_10', 'Q7_Part_11', 'Q7_Part_12']]
df = df.drop(df.index[0])

# Rename Columns 
df.columns = ['age','gender','country','education','job_title','years_coding','salary', 'data_science_schooling_1', 'data_science_schooling_2', 'data_science_schooling_3', 'data_science_schooling_4', 'data_science_schooling_5', 'data_science_schooling_6', 'data_science_schooling_7', 'data_science_schooling_8', 'data_science_schoolng_9', 'data_science_schooling_10', 'data_science_schooling_11', 'Languages_1', 'Languages_2', 'Languages_3', 'Languages_4', 'Languages_5', 'Languages_6', 'Languages_7', 'Languages_8', 'Languages_9', 'Languages_10', 'Languages_11', 'Languages_12']

# # Combine data sciences and languages columns 
df['Data_Science_Course'] = df[df.columns[8:18]].apply(
    lambda x: ';'.join(x.dropna().astype(str)),
    axis=1
)

df['Languages'] = df[df.columns[19:]].apply(
    lambda x: ';'.join(x.dropna().astype(str)),
    axis=1
)
```

```python
# Drop the individual schooling and languages columns 
df.drop(df.loc[:, 'data_science_schooling_1':'Languages_12'].columns, axis = 1, inplace = True)
df = df.dropna()
```
